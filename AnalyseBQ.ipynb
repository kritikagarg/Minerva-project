```python
from wordcloud import WordCloud
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()  
ps = PorterStemmer()
from string import punctuation
from collections import Counter
stoplist = stopwords.words('english') + list(punctuation)+['...',' ',''] 
```

### Top words in +ve documents


```python
d1 = sdata.loc[sdata['DocAvgTone'] > 0]
```

```python
len(d1)
```

```python
d1_counter = Counter([lemmatizer.lemmatize(i.lower()) for i in " ".join(d1['title']).split() if i.lower() not in stoplist]).most_common()
d1_count = pd.DataFrame(d1_counter)
d1_count.head(21)
```

```python
wordcloud = WordCloud().generate_from_frequencies(dict(d1_counter))
# Generate plot
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

### Top words in -ve documents

```python
d2 = sdata.loc[sdata['DocAvgTone'] < 0]
```

```python
len(d2)
```

```python
d2_counter = Counter([lemmatizer.lemmatize(i.lower()) for i in " ".join(d2['title'].astype(str)).split(" ") if i.lower() not in stoplist]).most_common()
d2_count = pd.DataFrame(d2_counter)
d2_count.head(21)
```

```python
wordcloud = WordCloud().generate_from_frequencies(dict(d2_counter))
# Generate plot
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
